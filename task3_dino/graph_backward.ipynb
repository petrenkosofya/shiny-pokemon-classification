```python
import torch
import torch.nn.functional as F

dtype = torch.float64
x1 = torch.tensor(3.0, dtype=dtype, requires_grad=True)
w1 = torch.tensor(4.0, dtype=dtype, requires_grad=True)
x2 = torch.tensor(10.0, dtype=dtype, requires_grad=True)
w2 = torch.tensor(5.0, dtype=dtype, requires_grad=True)
x3 = torch.tensor(1.0, dtype=dtype, requires_grad=True)
y  = torch.tensor(0.0, dtype=dtype)
```

```python
# Forward pass
a = x1 * w1
b = x2 / w2
c = b + x3
d = torch.maximum(c, a)
e = d - 28
f = torch.pow(torch.tensor(2.0, dtype=dtype), c)
g = a + e + f
p = torch.sigmoid(g)
loss = F.binary_cross_entropy(p, y)

print("Forward:")
print(f"a={a.item()}, \nb={b.item()}, \nc={c.item()}, \nd={d.item()}, \ne={e.item()}, \nf={f.item()}, \ng={g.item()}, \np={p.item()}")
print(f"loss(BCE)={loss.item()}")
```

```python
# Backward pass
loss.backward()

print("Backward:")
print(f"dloss/dx1={x1.grad.item()}, \ndloss/dw1={w1.grad.item()}, \ndloss/dx2={x2.grad.item()}, \ndloss/dw2={w2.grad.item()}, \ndloss/dx3={x3.grad.item()}")
```
